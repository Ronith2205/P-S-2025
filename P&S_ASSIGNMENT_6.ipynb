```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Question 1: Generate normal distribution and plot histogram
mean_1 = 60
std_dev_1 = 12
num_samples = 1000

data = np.random.normal(mean_1, std_dev_1, num_samples)

plt.figure(figsize=(10,5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

# Plot PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean_1, std_dev_1)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

plt.title("Histogram and PDF of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()

# Question 2: Calculate and visualize probabilities for calorie intake
mean_2 = 30
std_dev_2 = 5

def calculate_probability(value, mean, std_dev, condition):
    if condition == "greater":
        return 1 - stats.norm.cdf(value, mean, std_dev)
    elif condition == "less":
        return stats.norm.cdf(value, mean, std_dev)
    elif condition == "between":
        return stats.norm.cdf(value[1], mean, std_dev) - stats.norm.cdf(value[0], mean, std_dev)

# Probabilities
prob_greater_19 = calculate_probability(19, mean_2, std_dev_2, "greater")
prob_less_19 = calculate_probability(19, mean_2, std_dev_2, "less")
prob_between_19_35 = calculate_probability([19, 35], mean_2, std_dev_2, "between")

print(f"P(X >= 19) = {prob_greater_19:.4f}")
print(f"P(X <= 19) = {prob_less_19:.4f}")
print(f"P(19 <= X <= 35) = {prob_between_19_35:.4f}")

# Visualization
x = np.linspace(mean_2 - 4*std_dev_2, mean_2 + 4*std_dev_2, 1000)
y = stats.norm.pdf(x, mean_2, std_dev_2)

plt.figure(figsize=(10,5))
plt.plot(x, y, label='PDF', color='black')

# Shade areas for probability calculations
x_fill_1 = np.linspace(19, mean_2 + 4*std_dev_2, 100)
y_fill_1 = stats.norm.pdf(x_fill_1, mean_2, std_dev_2)
plt.fill_between(x_fill_1, y_fill_1, alpha=0.3, color='red', label='P(X >= 19)')

x_fill_2 = np.linspace(mean_2 - 4*std_dev_2, 19, 100)
y_fill_2 = stats.norm.pdf(x_fill_2, mean_2, std_dev_2)
plt.fill_between(x_fill_2, y_fill_2, alpha=0.3, color='blue', label='P(X <= 19)')

x_fill_3 = np.linspace(19, 35, 100)
y_fill_3 = stats.norm.pdf(x_fill_3, mean_2, std_dev_2)
plt.fill_between(x_fill_3, y_fill_3, alpha=0.3, color='green', label='P(19 <= X <= 35)')

plt.title("Probability Distribution of Daily Calorie Intake")
plt.xlabel("Calories")
plt.ylabel("Density")
plt.legend()
plt.show()
```
