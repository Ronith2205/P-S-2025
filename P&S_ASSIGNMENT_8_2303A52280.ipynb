```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr


daily_calories = np.array([2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000])
exercise_duration = np.array([60, 50, 30, 90, 45, 20, 75, 95, 15, 10])
bmi = np.array([25, 26, 28, 23, 27, 30, 24, 22, 31, 33])



def create_scatter_plot(x, y, xlabel, ylabel, title):
    plt.scatter(x, y, color='blue', edgecolors='black')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True)
    plt.show()


create_scatter_plot(daily_calories, bmi, 'Daily Calorie Intake', 'BMI', 'Daily Calorie Intake vs. BMI')


create_scatter_plot(exercise_duration, bmi, 'Exercise Duration (min)', 'BMI', 'Exercise Duration vs. BMI')


create_scatter_plot(daily_calories, exercise_duration, 'Daily Calorie Intake', 'Exercise Duration (min)', 'Daily Calorie Intake vs. Exercise Duration')



def calculate_correlation(x, y, label_x, label_y):
    correlation, _ = pearsonr(x, y)
    print(f'Pearson correlation coefficient between {label_x} and {label_y}: {correlation:.2f}')


calculate_correlation(daily_calories, bmi, 'Daily Calorie Intake', 'BMI')
calculate_correlation(exercise_duration, bmi, 'Exercise Duration', 'BMI')
calculate_correlation(daily_calories, exercise_duration, 'Daily Calorie Intake', 'Exercise Duration')


print("\nDiscussion:")
print("- If the correlation between Daily Calorie Intake and BMI is positive, it suggests that consuming more calories may lead to a higher BMI.")
print("- If the correlation between Exercise Duration and BMI is negative, it suggests that more exercise might be associated with lower BMI.")
print("- The relationship between Daily Calorie Intake and Exercise Duration helps understand lifestyle patterns.")
```
