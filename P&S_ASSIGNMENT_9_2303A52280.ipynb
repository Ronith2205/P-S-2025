```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, ttest_ind


daily_calories = [2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000]
exercise_duration = [60, 50, 30, 90, 45, 20, 75, 95, 15, 10]
bmi = [25, 26, 28, 23, 27, 30, 24, 22, 31, 33]


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(x=daily_calories, y=bmi)
plt.xlabel("Daily Calorie Intake")
plt.ylabel("BMI")
plt.title("Daily Calorie Intake vs. BMI")

plt.subplot(1, 3, 2)
sns.scatterplot(x=exercise_duration, y=bmi)
plt.xlabel("Exercise Duration (minutes)")
plt.ylabel("BMI")
plt.title("Exercise Duration vs. BMI")

plt.subplot(1, 3, 3)
sns.scatterplot(x=daily_calories, y=exercise_duration)
plt.xlabel("Daily Calorie Intake")
plt.ylabel("Exercise Duration (minutes)")
plt.title("Daily Calorie Intake vs. Exercise Duration")

plt.tight_layout()
plt.show()


corr_calories_bmi, _ = pearsonr(daily_calories, bmi)
corr_exercise_bmi, _ = pearsonr(exercise_duration, bmi)
corr_calories_exercise, _ = pearsonr(daily_calories, exercise_duration)

print(f"Pearson Correlation between Daily Calorie Intake and BMI: {corr_calories_bmi:.2f}")
print(f"Pearson Correlation between Exercise Duration and BMI: {corr_exercise_bmi:.2f}")
print(f"Pearson Correlation between Daily Calorie Intake and Exercise Duration: {corr_calories_exercise:.2f}")


if corr_calories_bmi > 0:
    print("Higher calorie intake is associated with a higher BMI.")
else:
    print("Higher calorie intake is not strongly associated with BMI.")

if corr_exercise_bmi < 0:
    print("More exercise tends to be linked with a lower BMI.")
else:
    print("Exercise duration does not significantly affect BMI.")



def perform_ttest(group1, group2, label):
    stat, p_value = ttest_ind(group1, group2, equal_var=False)
    print(f"T-test for {label}:")
    print(f"t-statistic: {stat:.2f}, p-value: {p_value:.4f}")
    if p_value < 0.05:
        print("There is a significant difference between the two groups.")
    else:
        print("There is no significant difference between the two groups.")
    print("-" * 50)


group_A_water = [2.8, 3.1, 3.0, 2.5, 3.2, 3.0, 3.3, 2.7, 2.9, 3.0]
group_B_water = [1.5, 1.7, 1.6, 1.9, 2.0, 1.8, 1.7, 1.6, 1.8, 1.5]
perform_ttest(group_A_water, group_B_water, "Daily Water Intake")


group_A_exercise = [7, 8, 6, 9, 10, 8, 7, 6, 8, 9]
group_B_exercise = [3, 4, 2, 3, 5, 4, 3, 2, 4, 3]
perform_ttest(group_A_exercise, group_B_exercise, "Weekly Exercise Hours")


group_X_reading = [4, 5, 6, 3, 5, 6, 4, 5, 7, 6]
group_Y_reading = [1, 2, 3, 2, 1, 4, 2, 1, 3, 2]
perform_ttest(group_X_reading, group_Y_reading, "Weekly Reading Hours")


```
